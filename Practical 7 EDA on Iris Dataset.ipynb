```python
plt.scatter(df['sepal.length'], df['sepal.width'],color="blue", edgecolor="red")
plt.title('Sepal length vs Sepal width')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.grid()
plt.show()
```

```python
dfset = df[['sepal.length','sepal.width','petal.length', 'petal.width',]]
sns.pairplot(dfset)
plt.show()
```

```python
sns.distplot(df["petal.width"], color="green")
plt.show()
```

```python
sns.distplot(df["sepal.width"], color="green")
plt.xlabel("Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(15,8),dpi=300)
sns.set_style("whitegrid")
sns.boxplot(x='variety', y='petal.length',data=df)
plt.xlabel('Variety')
plt.ylabel('Petal Length')
plt.show()
```

```python
plt.figure(figsize=(15,8),dpi=300)
sns.set_style("whitegrid")
sns.boxplot(x='petal.length', y='variety',data=df)
plt.ylabel('Variety')
plt.xlabel('Petal Length')
plt.show()
```
