```python
#drop_duplicates() method will return a copy of your DataFrame with duplicates removed from it.

temp_df2 = temp_df1.drop_duplicates()

temp_df2.shape
temp_df2
```

```python
#an important argument for drop_duplicates() is keep, which has three possible options:

#1. first:(default) Drop duplicates except for the first occurrence.
#2. last: drop duplicates except for the last occurrence.
#3. false: drop all duplicates.

#Aslso, using inplace = True will modify the DataFrame object in place.
```

```python
temp_df1.drop_duplicates(inplace=True, keep='first')
temp_df1.shape
```

```python
temp_df1.drop_duplicates(inplace=True, keep='last')
temp_df1.shape
```

```python
#Q5. Replace each column's Null values with either a constant or mean value.
df['country'].fillna("No Country", inplace=True) #replacing null values in country with "No country"
df['designation'].fillna("NA", inplace=True) #replacing null values in designation with "NA"

#replacing null values of price with the mean of price column
df['price'].fillna(df['price'].mean(), inplace= True)
df['province'].fillna("NA", inplace=True)
df['region_1'].fillna("NA", inplace=True)
df['region_2'].fillna("NA", inplace=True)
df['taster_name'].fillna("Unknown", inplace=True)
df['taster_twitter_handle'].fillna("Unknown", inplace=True)
df
```
