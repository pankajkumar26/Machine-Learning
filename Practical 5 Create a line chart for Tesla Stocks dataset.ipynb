# Demonstrate a Line chart using Tesla Stocks High Price dataset for 2016-2018.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataset : Tesla Stock Price Dataset
```

```python
stock_data = pd.read_csv("D:/ML dataset/tesla-stock-price.csv")
stock_data
```

```python
stock_data.head()
```

```python
# for ignoring all random warnings in the jupyter file
import warnings
warnings.filterwarnings('ignore')
```

```python
stock_data['date'] = pd.to_datetime(stock_data['date']) #converts date into datetime format with hyphens instead of slashes
stock_data.head()
```

```python
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_axes([0,0,1,1])
ax1.set_title('Tesla Stocks High Stock Prices')
ax1.plot(stock_data['date'],stock_data['high'],color='red')
ax1.plot(stock_data['date'],stock_data['low'],color='black')
plt.xticks(rotation=90)
ax1.grid()
plt.show()
```
